```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
website="https://ackodrive.com/collection/honda-cars/"
```

```python
response=requests.get(website)
```

```python
response.status_code
```

```python
# We need to create a soup object
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
results= soup.find_all('div',{'class':"styles__CarListWrapper-sc-6fe11a0-2 gwrxnw"})
```

```python
results
```

```python
len(results)
```

```python
morecar=soup.find_all('div',{"data-testid":"listingcardesktop",'class':"styles__Wrapper-sc-ad1b3a08-0 iTGjuL"})
```

```python
len(morecar)
```

```python
i = 0
data = []

while i < len(morecar):
    car = morecar[i]
    try:
        name = car.find('h2', {'class': 'styles__ModelTitle-sc-a6403e05-4 ARHdK'})
        name = name.get_text(strip=True) if name else "N/A"

        specs = car.find_all('div', {'class': 'styles__VariantInfo-sc-a6403e05-12 bkhKsh'})
        fuel = car.find('p',{'data-testid':"car_variant_fuel_type",'class':"styles__ParaWithoutMargins-sc-a6403e05-34 iAPjDz"})
        fuel = fuel.get_text(strip=True) if fuel else "N/A"
        transmission = car.find('p',{'data-testid':"car_variant_transmission",'class':"styles__ParaWithoutMargins-sc-a6403e05-34 iAPjDz"})
        transmission = transmission.get_text(strip=True) if transmission else "N/A"
        Availability = car.find('p',{'class':"styles__ParaWithoutMargins-sc-a6403e05-34 iAPjDz"})
        Availability = Availability.get_text(strip=True) if Availability else "N/A"
        price = car.find('div', {'class': 'styles__Price-sc-a6403e05-18 bsWAfs'})
        price = price.get_text(strip=True) if price else "N/A"

        data.append({
            "Car Name": name,
            "Fuel Type": fuel,
            "Transmission": transmission,
            "Availability": Availability,
            "Price": price
        })
    except Exception as e:
        print(f"Error on index {i}: {e}")
    
    i += 1

# Save to CSV
df = pd.DataFrame(data)
df.to_csv("honda_cars.csv", index=False)
print("✅ Data saved to honda_cars.csv")
```
